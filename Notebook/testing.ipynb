```python
import joblib
import os, sys
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import tabulate
```

```python
X_test = pd.read_csv(r'C:\Fatigue_Level\Data\Engineered_Data\X_test_transformed.csv')
y_test = pd.read_csv(r'C:\Fatigue_Level\Data\Engineered_Data\y_test.csv')
```

```python
model_path = r'C:\Fatigue_Level\Models\All_models'

decision_tree = os.path.join(model_path, 'Decision_tree.joblib')

decision_tree = joblib.load(decision_tree)
```

```python
model_path = r'C:\Fatigue_Level\Models\All_models'

random_forest = os.path.join(model_path, 'Random_forest.joblib')

random_forest = joblib.load(random_forest)
```

```python
model_path = r'C:\Fatigue_Level\Models\All_models'

XGboost_Classifier = os.path.join(model_path, 'XGboost_Classifier.joblib')

XGboost_Classifier = joblib.load(XGboost_Classifier)
```

```python
dt_pred = decision_tree.predict(X_test)
```

```python
dt_accuracy = accuracy_score(y_test, dt_pred)
dt_precision = precision_score(y_test, dt_pred, average='weighted')
dt_recall = recall_score(y_test, dt_pred, average='weighted')
dt_f1 = f1_score(y_test, dt_pred, average='weighted')
```

Random Forest

```python
rf_pred = random_forest.predict(X_test)
```

```python
rf_accuracy = accuracy_score(y_test, rf_pred)
rf_precision = precision_score(y_test, rf_pred, average='weighted')
rf_recall = recall_score(y_test, rf_pred, average='weighted')
rf_f1 = f1_score(y_test, rf_pred, average='weighted')
```

XGBoost Classifiation

```python
xgb = XGboost_Classifier.predict(X_test)
```

```python
xgb_accuracy = accuracy_score(y_test, xgb)
xgb_precision = precision_score(y_test, xgb, average='weighted')
xgb_recall = recall_score(y_test, xgb, average='weighted')
xgb_f1 = f1_score(y_test, xgb, average='weighted')
```

```python
metrics = [
    ['Decision Tree Classifier', dt_accuracy, dt_precision, dt_recall, dt_f1],
    ['Random Forest Classifier', rf_accuracy, rf_precision, rf_recall, rf_f1],
    ['XGBoost', xgb_accuracy, xgb_precision, xgb_recall, xgb_f1]
]

headers = ['Accuracy Score', 'Precision Score', 'Recall Score', 'F1 Score']

print(tabulate.tabulate(metrics, headers, floatfmt='.5f', tablefmt='orgtbl'))
```

```python
print(classification_report(y_test, xgb))
```
